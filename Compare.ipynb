```python
import pandas as pd

# Load your Excel file
file_path = r"C:\Users\vshankar\Downloads\HeidelbergMaterialsGroup_AECOHack_DataPackage\HUN Sales Database_YTD Aug25.xlsx"  # Replace with your actual file path
df = pd.read_excel(file_path)

# Display the first few rows to verify the data
print("Data preview:")
print(df.head())
print("\nColumn names:")
print(df.columns.tolist())

# ALGORITHM: Group by Month and Concrete type, then create truck pools
print("\n" + "="*80)
print("TRUCK POOLING ALGORITHM")
print("="*80)

# Constants for truck pooling
MAX_TRUCK_CAPACITY = 12  # Maximum m3 per truck
DISTANCE_THRESHOLD = 50  # km threshold for pooling
SMALL_ORDER_THRESHOLD = 3  # m3
MEDIUM_ORDER_THRESHOLD = 7  # m3

# Group by Month and Concrete type
grouped = df.groupby(['Month', 'Concrete type']).agg({
    'm3': 'sum',
    'Distance': 'mean'  # Average distance for the group
}).reset_index()

grouped.columns = ['Month', 'Concrete type', 'Total_m3', 'Avg_Distance']
grouped = grouped.sort_values(['Month', 'Concrete type'])

print("\n--- Orders by Month and Concrete Type ---")
print(grouped.to_string(index=False))

# CHANGED: Create truck pooling logic
print("\n" + "="*80)
print("TRUCK POOL RECOMMENDATIONS")
print("="*80)

truck_pools = []

for month in grouped['Month'].unique():
    month_data = grouped[grouped['Month'] == month].sort_values('Total_m3')
    print(f"\n--- MONTH: {month} ---")
    
    pools = []
    current_pool = []
    current_capacity = 0
    
    for _, row in month_data.iterrows():
        concrete_type = row['Concrete type']
        m3 = row['Total_m3']
        distance = row['Avg_Distance']
        
        # Check if we can add to current pool
        if current_capacity == 0:
            # Start new pool
            current_pool.append({
                'concrete_type': concrete_type,
                'm3': m3,
                'distance': distance
            })
            current_capacity = m3
        elif (current_capacity + m3 <= MAX_TRUCK_CAPACITY and 
              all(abs(distance - item['distance']) < DISTANCE_THRESHOLD for item in current_pool)):
            # Can add to current pool (same distance range and within capacity)
            current_pool.append({
                'concrete_type': concrete_type,
                'm3': m3,
                'distance': distance
            })
            current_capacity += m3
        else:
            # Cannot add - save current pool and start new one
            pools.append(current_pool)
            current_pool = [{
                'concrete_type': concrete_type,
                'm3': m3,
                'distance': distance
            }]
            current_capacity = m3
    
    # Add last pool
    if current_pool:
        pools.append(current_pool)
    
    # Print pools for this month
    for pool_num, pool in enumerate(pools, 1):
        total_m3 = sum(item['m3'] for item in pool)
        avg_distance = sum(item['distance'] for item in pool) / len(pool)
        
        print(f"\n  Truck Pool #{pool_num} (Total: {total_m3} m3 | Avg Distance: {avg_distance:.1f} km)")
        for item in pool:
            print(f"    - {item['concrete_type']}: {item['m3']} m3 @ {item['distance']:.1f} km")
        
        # Truck size recommendation
        if total_m3 <= 7:
            truck_size = "7 m3"
        elif total_m3 <= 12:
            truck_size = "12 m3"
        else:
            truck_size = f"SPLIT INTO MULTIPLE TRUCKS"
        
        print(f"    → Recommended Truck Size: {truck_size}")
        
        truck_pools.append({
            'Month': month,
            'Pool': pool_num,
            'Total_m3': total_m3,
            'Truck_Size': truck_size,
            'Avg_Distance': avg_distance
        })

# Summary of all pools
print("\n" + "="*80)
print("SUMMARY OF ALL TRUCK POOLS")
print("="*80)
summary_df = pd.DataFrame(truck_pools)
print(summary_df.to_string(index=False))

# Overall statistics
print("\n" + "="*80)
print("EFFICIENCY METRICS")
print("="*80)
print(f"Total Pools Created: {len(truck_pools)}")
print(f"Average m3 per Pool: {summary_df['Total_m3'].mean():.2f}")
print(f"Pools using 7 m3 truck: {len(summary_df[summary_df['Truck_Size'] == '7 m3'])}")
print(f"Pools using 12 m3 truck: {len(summary_df[summary_df['Truck_Size'] == '12 m3'])}")
```

```python
import pandas as pd

# Load your Excel file
file_path = r"C:\Users\vshankar\Downloads\HeidelbergMaterialsGroup_AECOHack_DataPackage\HUN Sales Database_YTD Aug25.xlsx"  # Replace with your actual file path
df = pd.read_excel(file_path)

# Display the first few rows to verify the data
print("Data preview:")
print(df.head())
print("\nColumn names:")
print(df.columns.tolist())

# ALGORITHM: Group by Month and Concrete type, then create truck pools
print("\n" + "="*80)
print("TRUCK POOLING ALGORITHM")
print("="*80)

# Constants for truck pooling
MAX_TRUCK_CAPACITY = 12  # Maximum m3 per truck
DISTANCE_THRESHOLD = 50  # km threshold for pooling
SMALL_ORDER_THRESHOLD = 3  # m3
MEDIUM_ORDER_THRESHOLD = 7  # m3

# Group by Month and Concrete type
grouped = df.groupby(['Month', 'Concrete type']).agg({
    'm3': 'sum',
    'Distance': 'mean'  # Average distance for the group
}).reset_index()

grouped.columns = ['Month', 'Concrete type', 'Total_m3', 'Avg_Distance']
grouped = grouped.sort_values(['Month', 'Concrete type'])

print("\n--- Orders by Month and Concrete Type ---")
print(grouped.to_string(index=False))

# CHANGED: Create truck pooling logic
print("\n" + "="*80)
print("TRUCK POOL RECOMMENDATIONS")
print("="*80)

truck_pools = []

for month in grouped['Month'].unique():
    month_data = grouped[grouped['Month'] == month].sort_values('Total_m3')
    print(f"\n--- MONTH: {month} ---")
    
    # CHANGED: POOLING CRITERIA - DISTANCE (PRIMARY), CONCRETE TYPE (SECONDARY), M3 (CAPACITY)
    # Step 1: Group by distance first (within threshold)
    distance_groups = []
    for _, row in month_data.iterrows():
        concrete_type = row['Concrete type']
        m3 = row['Total_m3']
        distance = row['Avg_Distance']
        
        # Try to add to existing distance group
        added = False
        for group in distance_groups:
            # Check if distance is within threshold of group
            group_avg_distance = sum(item['distance'] for item in group) / len(group)
            if abs(distance - group_avg_distance) < DISTANCE_THRESHOLD:
                group.append({
                    'concrete_type': concrete_type,
                    'm3': m3,
                    'distance': distance
                })
                added = True
                break
        
        # If not added to any group, create new distance group
        if not added:
            distance_groups.append([{
                'concrete_type': concrete_type,
                'm3': m3,
                'distance': distance
            }])
    
    # Step 2: Within each distance group, group by Concrete type
    pools = []
    for distance_group in distance_groups:
        # Group by concrete type within this distance group
        concrete_type_groups = {}
        for item in distance_group:
            ct = item['concrete_type']
            if ct not in concrete_type_groups:
                concrete_type_groups[ct] = []
            concrete_type_groups[ct].append(item)
        
        # Step 3: Pool by m3 capacity within each concrete type group
        for concrete_type, items in concrete_type_groups.items():
            items.sort(key=lambda x: x['m3'])
            
            current_pool = []
            current_capacity = 0
            
            for item in items:
                m3 = item['m3']
                
                # Check if we can add to current pool (by m3 capacity)
                if current_capacity == 0:
                    # Start new pool
                    current_pool.append(item)
                    current_capacity = m3
                elif current_capacity + m3 <= MAX_TRUCK_CAPACITY:
                    # Can add to current pool (same distance, same concrete type, within capacity)
                    current_pool.append(item)
                    current_capacity += m3
                else:
                    # Cannot add - save current pool and start new one
                    pools.append(current_pool)
                    current_pool = [item]
                    current_capacity = m3
            
            # Add last pool
            if current_pool:
                pools.append(current_pool)
    
    # Print pools for this month
    for pool_num, pool in enumerate(pools, 1):
        total_m3 = sum(item['m3'] for item in pool)
        avg_distance = sum(item['distance'] for item in pool) / len(pool)
        
        print(f"\n  Truck Pool #{pool_num} (Total: {total_m3} m3 | Avg Distance: {avg_distance:.1f} km)")
        for item in pool:
            print(f"    - {item['concrete_type']}: {item['m3']} m3 @ {item['distance']:.1f} km")
        
        # Truck size recommendation
        if total_m3 <= 7:
            truck_size = "7 m3"
        elif total_m3 <= 12:
            truck_size = "12 m3"
        else:
            truck_size = f"SPLIT INTO MULTIPLE TRUCKS"
        
        print(f"    → Recommended Truck Size: {truck_size}")
        
        truck_pools.append({
            'Month': month,
            'Pool': pool_num,
            'Total_m3': total_m3,
            'Truck_Size': truck_size,
            'Avg_Distance': avg_distance
        })

# Summary of all pools
print("\n" + "="*80)
print("SUMMARY OF ALL TRUCK POOLS")
print("="*80)
summary_df = pd.DataFrame(truck_pools)
print(summary_df.to_string(index=False))

# Overall statistics
print("\n" + "="*80)
print("EFFICIENCY METRICS")
print("="*80)
print(f"Total Pools Created: {len(truck_pools)}")
print(f"Average m3 per Pool: {summary_df['Total_m3'].mean():.2f}")
print(f"Pools using 7 m3 truck: {len(summary_df[summary_df['Truck_Size'] == '7 m3'])}")
print(f"Pools using 12 m3 truck: {len(summary_df[summary_df['Truck_Size'] == '12 m3'])}")
```

```python
# CHANGED: Install required libraries first (run this cell)
import subprocess
import sys

subprocess.check_call([sys.executable, "-m", "pip", "install", "geopy"])

import pandas as pd
from geopy.geocoders import Nominatim
from geopy.distance import geodesic
import time

# Load your Excel file
file_path = r"C:\Users\vshankar\Downloads\HeidelbergMaterialsGroup_AECOHack_DataPackage\HUN Sales Database_YTD Aug25.xlsx"  # Replace with your actual file path
df = pd.read_excel(file_path)

# CHANGED: Initialize geocoder for address to coordinates conversion
geolocator = Nominatim(user_agent="concrete_pooling_app")

# Display the first few rows to verify the data
print("Data preview:")
print(df.head())
print("\nColumn names:")
print(df.columns.tolist())

# CHANGED: Convert factory location to coordinates using geopy
print("Getting factory coordinates...")
factory_location = geolocator.geocode("Papírgyári út 42.-46., 2400 Dunaújváros, Hungary")
FACTORY_LOCATION = {
    'name': 'Papírgyári út 42.-46., 2400 Dunaújváros',
    'lat': factory_location.latitude,
    'lon': factory_location.longitude
}

# Constants for truck pooling
MAX_TRUCK_CAPACITY = 12  # Maximum m3 per truck
DISTANCE_THRESHOLD = 50  # km threshold for pooling
MAX_ROUTE_DISTANCE = 200  # Maximum total distance for one route (km)

# CHANGED: Calculate actual distance from factory to each location using geodesic
def get_distance_from_factory(location_name):
    """CHANGED: Get actual distance using geopy geodesic calculation"""
    if location_name in location_coords:
        loc = location_coords[location_name]
        factory_coords = (FACTORY_LOCATION['lat'], FACTORY_LOCATION['lon'])
        location_point = (loc['lat'], loc['lon'])
        distance_km = geodesic(factory_coords, location_point).kilometers
        return distance_km
    return 0

# Add calculated distances to dataframe
df['Calculated_Distance'] = df['Location'].apply(get_distance_from_factory)

print("\n" + "="*80)
print("TRUCK POOLING ALGORITHM WITH LOCATION-BASED ROUTING")
print("="*80)
print(f"Factory Location: {FACTORY_LOCATION['name']}")
print(f"Distance Threshold: {DISTANCE_THRESHOLD} km")
print(f"Max Route Distance: {MAX_ROUTE_DISTANCE} km")
print(f"Max Truck Capacity: {MAX_TRUCK_CAPACITY} m3")

# CHANGED: Geocode all unique locations to get coordinates
print("\nGeocoding all customer locations (this may take a moment)...")
unique_locations = df['Location'].unique()
location_coords = {}

for loc in unique_locations:
    try:
        print(f"  Geocoding: {loc}...")
        # Add Hungary to improve accuracy
        search_address = f"{loc}, Hungary"
        geo_loc = geolocator.geocode(search_address)
        if geo_loc:
            location_coords[loc] = {
                'lat': geo_loc.latitude,
                'lon': geo_loc.longitude,
                'address': geo_loc.address
            }
            print(f"    ✓ Found at {geo_loc.latitude}, {geo_loc.longitude}")
        else:
            print(f"    ✗ Could not geocode")
        time.sleep(1)  # Rate limiting for API
    except Exception as e:
        print(f"    ✗ Error: {e}")

print("\n--- Orders by Month, Location, and Concrete Type ---")
# Group to show all orders
order_summary = df.groupby(['Month', 'Location', 'Concrete type']).agg({
    'm3': 'sum',
    'Distance': 'first'  # Distance from factory
}).reset_index()
order_summary.columns = ['Month', 'Location', 'Concrete type', 'Total_m3', 'Distance_from_Factory']
order_summary = order_summary.sort_values(['Month', 'Distance_from_Factory', 'Location'])
print(order_summary.to_string(index=False))

# CHANGED: Create advanced routing pools
print("\n" + "="*80)
print("OPTIMIZED TRUCK ROUTES (Amazon-style Routing)")
print("="*80)

truck_routes = []
route_counter = 0

for month in df['Month'].unique():
    month_data = df[df['Month'] == month].copy()
    print(f"\n{'='*80}")
    print(f"MONTH: {month}")
    print(f"{'='*80}")
    
    # Sort by distance from factory (nearest first - TSP optimization)
    month_data = month_data.sort_values('Distance')
    
    # Group by location and concrete type first
    location_groups = month_data.groupby(['Location', 'Concrete type']).agg({
        'm3': 'sum',
        'Distance': 'first'
    }).reset_index()
    location_groups = location_groups.sort_values('Distance')
    
    print(f"\n--- Orders by Location (sorted by distance from factory) ---")
    print(location_groups.to_string(index=False))
    
    # CHANGED: Create routes using clustering by distance and location
    routes = []
    unassigned = location_groups.copy()
    
    while len(unassigned) > 0:
        route_counter += 1
        current_route = []
        current_capacity = 0
        current_distance = 0
        
        # Start with the nearest unassigned location
        first_idx = unassigned.index[0]
        first_row = unassigned.loc[first_idx]
        
        current_route.append({
            'location': first_row['Location'],
            'concrete_type': first_row['Concrete type'],
            'm3': first_row['m3'],
            'distance_from_factory': first_row['Distance']
        })
        current_capacity = first_row['m3']
        current_distance = first_row['Distance']
        unassigned = unassigned.drop(first_idx)
        
        # Try to add nearby locations to this route
        while len(unassigned) > 0:
            # Find next best order (nearby, same or similar distance, fits capacity)
            best_idx = None
            best_distance_diff = float('inf')
            
            for idx, row in unassigned.iterrows():
                location_distance = row['Distance']
                distance_diff = abs(location_distance - current_distance)
                
                # CHANGED: Criteria for adding to route:
                # 1. Within distance threshold
                # 2. Fits in truck capacity
                # 3. Prefer nearest distance
                if (distance_diff <= DISTANCE_THRESHOLD and 
                    current_capacity + row['m3'] <= MAX_TRUCK_CAPACITY and
                    distance_diff < best_distance_diff):
                    best_idx = idx
                    best_distance_diff = distance_diff
            
            if best_idx is not None:
                row = unassigned.loc[best_idx]
                current_route.append({
                    'location': row['Location'],
                    'concrete_type': row['Concrete type'],
                    'm3': row['m3'],
                    'distance_from_factory': row['Distance']
                })
                current_capacity += row['m3']
                unassigned = unassigned.drop(best_idx)
            else:
                break
        
        routes.append(current_route)
    
    # Display routes
    for route_num, route in enumerate(routes, 1):
        total_m3 = sum(item['m3'] for item in route)
        avg_distance = sum(item['distance_from_factory'] for item in route) / len(route)
        max_distance = max(item['distance_from_factory'] for item in route)
        
        print(f"\n--- ROUTE #{route_counter} ---")
        print(f"Total Capacity Used: {total_m3}/{MAX_TRUCK_CAPACITY} m3")
        print(f"Average Distance: {avg_distance:.1f} km")
        print(f"Max Distance in Route: {max_distance:.1f} km")
        print(f"Orders in Route:")
        
        for idx, item in enumerate(route, 1):
            print(f"  {idx}. Location: {item['location']}")
            print(f"     Concrete Type: {item['concrete_type']}")
            print(f"     m3: {item['m3']}")
            print(f"     Distance from Factory: {item['distance_from_factory']:.1f} km")
        
        # Truck size recommendation
        if total_m3 <= 7:
            truck_size = "7 m3"
        elif total_m3 <= 12:
            truck_size = "12 m3"
        else:
            truck_size = "NEEDS SPLIT"
        
        print(f"  → Recommended Truck Size: {truck_size}")
        
        truck_routes.append({
            'Month': month,
            'Route': route_counter,
            'Total_m3': total_m3,
            'Truck_Size': truck_size,
            'Avg_Distance': avg_distance,
            'Max_Distance': max_distance,
            'Orders_Count': len(route)
        })

# Summary
print("\n" + "="*80)
print("ROUTING SUMMARY")
print("="*80)
summary_df = pd.DataFrame(truck_routes)
print(summary_df.to_string(index=False))

print("\n" + "="*80)
print("EFFICIENCY METRICS")
print("="*80)
print(f"Total Routes Created: {route_counter}")
print(f"Average m3 per Route: {summary_df['Total_m3'].mean():.2f}")
print(f"Routes using 7 m3 truck: {len(summary_df[summary_df['Truck_Size'] == '7 m3'])}")
print(f"Routes using 12 m3 truck: {len(summary_df[summary_df['Truck_Size'] == '12 m3'])}")
print(f"Average Distance per Route: {summary_df['Avg_Distance'].mean():.1f} km")
```

```python
import pandas as pd
from math import radians, cos, sin, asin, sqrt
import random
# Load your Excel file
file_path = r"C:\Users\vshankar\Downloads\HeidelbergMaterialsGroup_AECOHack_DataPackage\HUN Sales Database_YTD Aug25.xlsx"  # Replace with your actual file path
df = pd.read_excel(file_path)

print("Data preview:")
print(df.head())

# CHANGED: Define factory location
FACTORY_LOCATION = {
    'name': 'Papírgyári út 42.-46., 2400 Dunaújváros',
    'lat': 46.9667,
    'lon': 18.9333
}

# Constants for truck pooling
DISTANCE_THRESHOLD = 50  # km threshold for pooling

# CHANGED: TRUCK FLEET DEFINITION (like DHL, UPS, Amazon)
TRUCK_SIZES = {
    '7': {'capacity': 7, 'price_per_m3': 104.23},      # Medium van/truck
    '12': {'capacity': 12, 'price_per_m3': 28.57}      # Large truck
}

# CHANGED: PRICING STRUCTURE (by order size, no pooling)
PRICING_NO_POOL = {
    'less_than_3': 124.00,      # euros per m3
    '3_to_7': 104.23,           # euros per m3
    'more_than_7': 28.57        # euros per m3
}

print("\n" + "="*80)
print("FLEET MANAGEMENT OPTIMIZATION SYSTEM")
print("="*80)
print("\nAvailable Trucks:")
for truck_type, specs in TRUCK_SIZES.items():
    print(f"  {truck_type} m³ Truck: €{specs['price_per_m3']} per m³ (capacity: {specs['capacity']} m³)")

# CHANGED: Function to generate random locations based on Distance
random.seed(42)

def generate_random_location(distance_km):
    """Generate random coordinates at approximately 'distance_km' from factory"""
    angle = random.uniform(0, 360)
    angle_rad = radians(angle)
    lat_offset = (distance_km / 111) * cos(angle_rad)
    lon_offset = (distance_km / 111) * sin(angle_rad)
    new_lat = FACTORY_LOCATION['lat'] + lat_offset
    new_lon = FACTORY_LOCATION['lon'] + lon_offset
    return new_lat, new_lon

# Add generated coordinates
df['Generated_Lat'] = 0.0
df['Generated_Lon'] = 0.0
for idx, row in df.iterrows():
    if pd.notna(row['Distance']):
        lat, lon = generate_random_location(row['Distance'])
        df.at[idx, 'Generated_Lat'] = lat
        df.at[idx, 'Generated_Lon'] = lon

# CHANGED: Calculate cost per order BEFORE pooling
def calculate_order_cost(m3):
    """Calculate cost based on order size without pooling"""
    if m3 < 3:
        return m3 * PRICING_NO_POOL['less_than_3']
    elif m3 <= 7:
        return m3 * PRICING_NO_POOL['3_to_7']
    else:
        return m3 * PRICING_NO_POOL['more_than_7']

df['Cost_Before_Pooling'] = df['m3'].apply(calculate_order_cost)

print("\n" + "="*80)
print("ORDERS SUMMARY (BEFORE OPTIMIZATION)")
print("="*80)
before_summary = df.groupby('Month').agg({
    'm3': 'sum',
    'Cost_Before_Pooling': 'sum'
}).reset_index()
before_summary.columns = ['Month', 'Total_m3', 'Total_Cost_Before']
print(before_summary)

# CHANGED: ADVANCED FLEET OPTIMIZATION ENGINE
print("\n" + "="*80)
print("OPTIMIZING FLEET ROUTES (Month → Concrete Type → Distance → Truck Size)")
print("="*80)

truck_routes = []
route_counter = 0

for month in sorted(df['Month'].unique()):
    month_data = df[df['Month'] == month].copy()
    print(f"\n{'='*80}")
    print(f"MONTH: {month}")
    print(f"{'='*80}")
    
    # CHANGED: Step 1 - Group by Concrete Type
    for concrete_type in sorted(month_data['Concrete type'].unique()):
        concrete_data = month_data[month_data['Concrete type'] == concrete_type].copy()
        
        print(f"\n  Concrete Type: {concrete_type}")
        
        # CHANGED: Step 2 - Sort by Distance from plant
        concrete_data = concrete_data.sort_values('Distance').reset_index(drop=True)
        
        # CHANGED: Step 3 - Create distance zones (clusters)
        distance_zones = []
        current_zone = []
        last_distance = None
        
        for idx, row in concrete_data.iterrows():
            if last_distance is None or abs(row['Distance'] - last_distance) <= DISTANCE_THRESHOLD:
                current_zone.append(idx)
                last_distance = row['Distance']
            else:
                if current_zone:
                    distance_zones.append(current_zone)
                current_zone = [idx]
                last_distance = row['Distance']
        
        if current_zone:
            distance_zones.append(current_zone)
        
        # CHANGED: Step 4 - Optimize truck assignment within each distance zone
        for zone_num, zone_indices in enumerate(distance_zones, 1):
            zone_data = concrete_data.loc[zone_indices].copy()
            total_m3 = zone_data['m3'].sum()
            avg_distance = zone_data['Distance'].mean()
            
            print(f"    Distance Zone {zone_num} (~{avg_distance:.1f}km):")
            print(f"      Orders: {len(zone_indices)}, Total m3: {total_m3}")
            
            # CHANGED: Step 5 - Select optimal truck size for this zone
            # Optimize by trying to fill trucks efficiently
            orders = zone_data['m3'].tolist()
            orders.sort(reverse=True)  # Sort largest first (bin packing algorithm)
            
            truck_assignments = []
            remaining_orders = orders.copy()
            
            # Greedy bin packing - try 12m3 truck first, then 7m3
            while remaining_orders:
                # Try to fill a 12m3 truck
                truck_12_load = []
                truck_12_capacity = 0
                
                for i in range(len(remaining_orders)):
                    if truck_12_capacity + remaining_orders[i] <= TRUCK_SIZES['12']['capacity']:
                        truck_12_load.append(remaining_orders[i])
                        truck_12_capacity += remaining_orders[i]
                
                if truck_12_load:
                    truck_assignments.append({'size': '12', 'load': truck_12_load, 'total': truck_12_capacity})
                    for item in truck_12_load:
                        remaining_orders.remove(item)
                else:
                    # If can't fill 12m3, try 7m3 truck
                    truck_7_load = []
                    truck_7_capacity = 0
                    
                    for i in range(len(remaining_orders)):
                        if truck_7_capacity + remaining_orders[i] <= TRUCK_SIZES['7']['capacity']:
                            truck_7_load.append(remaining_orders[i])
                            truck_7_capacity += remaining_orders[i]
                    
                    if truck_7_load:
                        truck_assignments.append({'size': '7', 'load': truck_7_load, 'total': truck_7_capacity})
                        for item in truck_7_load:
                            remaining_orders.remove(item)
                    else:
                        # Single order larger than available trucks
                        truck_assignments.append({'size': '12', 'load': [remaining_orders[0]], 'total': remaining_orders[0]})
                        remaining_orders.pop(0)
            
            # Calculate costs for this zone
            zone_cost_before = zone_data['Cost_Before_Pooling'].sum()
            zone_cost_after = 0
            
            for truck_num, truck in enumerate(truck_assignments, 1):
                truck_size = truck['size']
                truck_load = truck['total']
                cost = truck_load * TRUCK_SIZES[truck_size]['price_per_m3']
                zone_cost_after += cost
                
                route_counter += 1
                
                print(f"      Truck {truck_num} ({truck_size} m³): {len(truck['load'])} orders, {truck_load} m³ → €{cost:.2f}")
                
                truck_routes.append({
                    'Month': month,
                    'Route': route_counter,
                    'Concrete_Type': concrete_type,
                    'Distance_Zone': f"{avg_distance:.0f}km",
                    'Truck_Size': truck_size,
                    'Orders_Count': len(truck['load']),
                    'Total_m3': truck_load,
                    'Utilization': f"{(truck_load / TRUCK_SIZES[truck_size]['capacity'] * 100):.1f}%",
                    'Cost_Before': zone_cost_before / len(truck_assignments),
                    'Cost_After': cost,
                    'Route_Savings': (zone_cost_before / len(truck_assignments)) - cost
                })
            
            zone_savings = zone_cost_before - zone_cost_after
            zone_savings_pct = (zone_savings / zone_cost_before * 100) if zone_cost_before > 0 else 0
            
            print(f"      Zone Total - Before: €{zone_cost_before:.2f} | After: €{zone_cost_after:.2f} | Savings: €{zone_savings:.2f} ({zone_savings_pct:.1f}%)")

# CHANGED: Create comprehensive summary report
print("\n" + "="*80)
print("FLEET OPTIMIZATION SUMMARY")
print("="*80)
routes_df = pd.DataFrame(truck_routes)
print(routes_df[['Month', 'Route', 'Concrete_Type', 'Distance_Zone', 'Truck_Size', 
                  'Orders_Count', 'Total_m3', 'Utilization', 'Cost_After', 'Route_Savings']].to_string(index=False))

# CHANGED: Calculate total savings and metrics
print("\n" + "="*80)
print("TOTAL OPTIMIZATION RESULTS")
print("="*80)

total_cost_before = df['Cost_Before_Pooling'].sum()
total_cost_after = routes_df['Cost_After'].sum()
total_savings = total_cost_before - total_cost_after
total_savings_percent = (total_savings / total_cost_before * 100) if total_cost_before > 0 else 0

print(f"\nTotal Cost (Before Optimization): €{total_cost_before:.2f}")
print(f"Total Cost (After Optimization):  €{total_cost_after:.2f}")
print(f"Total Savings:                    €{total_savings:.2f}")
print(f"Savings Percentage:               {total_savings_percent:.1f}%")

print(f"\nTotal Routes Required: {route_counter}")
print(f"Average Truck Utilization: {routes_df['Utilization'].apply(lambda x: float(x.rstrip('%'))).mean():.1f}%")

# Truck size breakdown
print("\n--- TRUCK FLEET BREAKDOWN ---")
truck_breakdown = routes_df.groupby('Truck_Size').agg({
    'Route': 'count',
    'Total_m3': 'sum',
    'Cost_After': 'sum',
    'Route_Savings': 'sum'
}).reset_index()
truck_breakdown.columns = ['Truck_Size', 'Count', 'Total_m3', 'Total_Cost', 'Total_Savings']
print(truck_breakdown.to_string(index=False))

# Monthly breakdown
print("\n--- MONTHLY OPTIMIZATION BREAKDOWN ---")
monthly_breakdown = routes_df.groupby('Month').agg({
    'Route': 'count',
    'Total_m3': 'sum',
    'Cost_After': 'sum',
    'Route_Savings': 'sum'
}).reset_index()
monthly_breakdown.columns = ['Month', 'Routes', 'Total_m3', 'Total_Cost_After', 'Total_Savings']

# Add before costs
for idx, month in enumerate(monthly_breakdown['Month']):
    month_before = before_summary[before_summary['Month'] == month]['Total_Cost_Before'].values
    if len(month_before) > 0:
        monthly_breakdown.at[idx, 'Cost_Before'] = month_before[0]

monthly_breakdown['Savings_Percent'] = ((monthly_breakdown['Cost_Before'] - monthly_breakdown['Total_Cost_After']) / monthly_breakdown['Cost_Before'] * 100).round(1)
print(monthly_breakdown[['Month', 'Routes', 'Cost_Before', 'Total_Cost_After', 'Total_Savings', 'Savings_Percent']].to_string(index=False))

print("\n" + "="*80)
print("✓ OPTIMIZATION COMPLETE - Ready for Fleet Deployment")
```

```python
# fleet_ilp_optimizer.py
# Full end-to-end script: ILP optimizer per distance-zone with greedy fallback.
# Requirements: pandas, openpyxl (for reading .xlsx). PuLP recommended for ILP:
#   pip install pandas openpyxl pulp

import pandas as pd
import math
import random
from math import radians, cos
import sys

# Try to import pulp; if missing, we'll fall back to greedy solver automatically.
PULP_AVAILABLE = True
try:
    import pulp
except Exception:
    PULP_AVAILABLE = False

# ---------------- User-editable settings ----------------
file_path = r"C:\Users\vshankar\Downloads\HeidelbergMaterialsGroup_AECOHack_DataPackage\HUN Sales Database_YTD Aug25.xlsx"
FACTORY_LOCATION = {'name': 'Papírgyári út 42.-46., 2400 Dunaújváros', 'lat': 46.9667, 'lon': 18.9333}
DISTANCE_THRESHOLD = 50  # zone threshold (km)

TRUCK_SIZES = {
    '7':  {'capacity': 7.0,  'price_per_m3': 104.23},
    '12': {'capacity': 12.0, 'price_per_m3': 28.57}
}

PRICING_NO_POOL = {
    'less_than_3': 124.00,
    '3_to_7': 104.23,
    'more_than_7': 28.57
}

PRICING_POOLED = {
    '7': 104.23,
    '12': 28.57
}

random.seed(42)
# -------------------------------------------------------

def generate_random_location(factory_lat, factory_lon, distance_km, seed=None):
    """Generate approximate lat/lon at ~distance_km from factory (longitude scaled by cos(lat))."""
    if seed is not None:
        random.seed(seed)
    angle = random.uniform(0, 360)
    angle_rad = radians(angle)
    lat_offset = (distance_km / 111.0) * math.cos(angle_rad)
    lon_km_per_deg = 111.0 * math.cos(radians(factory_lat))
    if abs(lon_km_per_deg) < 1e-9:
        lon_km_per_deg = 1e-9
    lon_offset = (distance_km * math.sin(angle_rad)) / lon_km_per_deg
    return factory_lat + lat_offset, factory_lon + lon_offset

def calculate_order_cost(m3):
    if m3 < 3:
        return m3 * PRICING_NO_POOL['less_than_3']
    elif m3 <= 7:
        return m3 * PRICING_NO_POOL['3_to_7']
    else:
        return m3 * PRICING_NO_POOL['more_than_7']

# ---------- Greedy FFD bin-packer (fallback) ----------
def greedy_pack(zone_df):
    """First-Fit Decreasing greedy packing.
       Returns list of truck_assignments like ILP solver."""
    orders = [{'order_id': row['order_id'], 'm3': float(row['m3'])} for _, row in zone_df.iterrows()]
    orders.sort(key=lambda x: x['m3'], reverse=True)
    truck_assignments = []
    # Use capacities ordered ascending for choosing smallest fitting truck when opening new one
    truck_sizes_sorted = sorted(TRUCK_SIZES.items(), key=lambda kv: kv[1]['capacity'])
    for o in orders:
        placed = False
        for t in truck_assignments:
            cap = TRUCK_SIZES[t['size']]['capacity']
            if t['total'] + o['m3'] <= cap + 1e-6:
                t['load'].append(o['order_id'])
                t['total'] += o['m3']
                placed = True
                break
        if not placed:
            # choose smallest truck that fits this single order
            chosen_size = None
            for size_key, specs in truck_sizes_sorted:
                if o['m3'] <= specs['capacity'] + 1e-6:
                    chosen_size = size_key
                    break
            if chosen_size is None:
                # order larger than any single truck: choose largest truck (will overflow unless splitting allowed)
                chosen_size = truck_sizes_sorted[-1][0]
            truck_assignments.append({
                'size': chosen_size,
                'load': [o['order_id']],
                'total': o['m3'],
                'description': f"Orders: 1 (id {o['order_id']})"
            })
    # Map to same structure used later by summary (size keys are strings)
    return truck_assignments

# ---------- ILP solver using PuLP ----------
def solve_zone_ilp(zone_df, truck_sizes, time_limit=60):
    """Formulate and solve integer linear program packing orders into trucks.
       Requires PuLP. Returns list of truck assignments and total_cost."""
    if not PULP_AVAILABLE:
        raise RuntimeError("PuLP not available. Install pulp to use ILP solver: pip install pulp")
    orders = list(zone_df['order_id'])
    m3 = {row['order_id']: float(row['m3']) for _, row in zone_df.iterrows()}
    T = len(orders)  # upper bound on trucks
    truck_indices = list(range(T))
    truck_types = list(truck_sizes.keys())

    model = pulp.LpProblem("Truck_Packing_Zone", pulp.LpMinimize)

    # Variables
    x = pulp.LpVariable.dicts('x', [(i,t) for i in orders for t in truck_indices], cat='Binary')
    y = pulp.LpVariable.dicts('y', [(t,k) for t in truck_indices for k in truck_types], cat='Binary')
    open_t = pulp.LpVariable.dicts('open', truck_indices, cat='Binary')
    z = pulp.LpVariable.dicts('z', [(i,t,k) for i in orders for t in truck_indices for k in truck_types], cat='Binary')

    # Constraints
    for i in orders:
        model += pulp.lpSum(x[(i,t)] for t in truck_indices) == 1

    for t in truck_indices:
        for i in orders:
            model += x[(i,t)] <= open_t[t]
        model += pulp.lpSum(y[(t,k)] for k in truck_types) <= 1
        model += pulp.lpSum(y[(t,k)] for k in truck_types) >= open_t[t]

        # capacity
        model += pulp.lpSum(m3[i] * x[(i,t)] for i in orders) <= pulp.lpSum(truck_sizes[k]['capacity'] * y[(t,k)] for k in truck_types)

    # linearize cost with z variables linking x and y
    for i in orders:
        for t in truck_indices:
            for k in truck_types:
                model += z[(i,t,k)] <= x[(i,t)]
                model += z[(i,t,k)] <= y[(t,k)]
                model += z[(i,t,k)] >= x[(i,t)] + y[(t,k)] - 1
            model += pulp.lpSum(z[(i,t,k)] for k in truck_types) == x[(i,t)]

    model += pulp.lpSum(truck_sizes[k]['price_per_m3'] * m3[i] * z[(i,t,k)]
                        for i in orders for t in truck_indices for k in truck_types)

    solver = pulp.PULP_CBC_CMD(msg=False, timeLimit=time_limit)
    res = model.solve(solver)

    if pulp.LpStatus[model.status] not in ('Optimal', 'Integer Feasible'):
        print("Warning: solver status:", pulp.LpStatus[model.status])

    truck_assignments = []
    for t in truck_indices:
        if pulp.value(open_t[t]) and pulp.value(open_t[t]) > 0.5:
            chosen_type = None
            for k in truck_types:
                if pulp.value(y[(t,k)]) and pulp.value(y[(t,k)]) > 0.5:
                    chosen_type = k
                    break
            assigned_orders = [i for i in orders if pulp.value(x[(i,t)]) > 0.5]
            total_load = sum(m3[i] for i in assigned_orders)
            truck_assignments.append({
                'size': chosen_type,
                'load': assigned_orders,
                'total': total_load,
                'description': f"Orders: {len(assigned_orders)}"
            })

    total_cost = pulp.value(model.objective)
    return truck_assignments, total_cost

# ---------------- Main process: read file, precompute, cluster, optimize ----------------

def main(use_ilp=True, ilp_time_limit=60):
    # Load Excel
    try:
        df = pd.read_excel(file_path)
    except Exception as e:
        print("Error reading Excel file. Check file_path. Exception:", e)
        sys.exit(1)

    # Ensure necessary columns exist; create order_id if missing
    if 'order_id' not in df.columns:
        df['order_id'] = df.index.astype(str)  # fallback

    required_cols = ['Distance', 'm3', 'Month', 'Concrete type']
    for c in required_cols:
        if c not in df.columns:
            print(f"ERROR: required column '{c}' not found in Excel. Please ensure your sheet contains it.")
            sys.exit(1)

    # Add generated coordinates
    df['Generated_Lat'] = 0.0
    df['Generated_Lon'] = 0.0
    for idx, row in df.iterrows():
        if pd.notna(row['Distance']):
            lat, lon = generate_random_location(FACTORY_LOCATION['lat'], FACTORY_LOCATION['lon'], row['Distance'])
            df.at[idx, 'Generated_Lat'] = lat
            df.at[idx, 'Generated_Lon'] = lon

    # Calculate cost before pooling
    df['Cost_Before_Pooling'] = df['m3'].apply(calculate_order_cost)

    before_summary = df.groupby('Month').agg({'m3': 'sum', 'Cost_Before_Pooling': 'sum'}).reset_index()
    before_summary.columns = ['Month', 'Total_m3', 'Total_Cost_Before']
    print("\nBefore optimization summary (by Month):")
    print(before_summary.to_string(index=False))

    truck_routes = []
    route_counter = 0

    for month in sorted(df['Month'].unique()):
        month_data = df[df['Month'] == month].copy()
        print(f"\nProcessing month: {month}")

        for concrete_type in sorted(month_data['Concrete type'].unique()):
            concrete_data = month_data[month_data['Concrete type'] == concrete_type].copy()
            concrete_data = concrete_data.sort_values('Distance').reset_index(drop=True)

            # build distance zones
            distance_zones = []
            current_zone = []
            last_distance = None
            for idx, row in concrete_data.iterrows():
                if last_distance is None or abs(row['Distance'] - last_distance) <= DISTANCE_THRESHOLD:
                    current_zone.append(idx)
                    last_distance = row['Distance']
                else:
                    if current_zone:
                        distance_zones.append(current_zone)
                    current_zone = [idx]
                    last_distance = row['Distance']
            if current_zone:
                distance_zones.append(current_zone)

            # per zone optim
            for zone_num, zone_indices in enumerate(distance_zones, 1):
                zone_data = concrete_data.loc[zone_indices].copy()
                # ensure order_id exists and is unique within zone
                zone_data = zone_data.reset_index(drop=True)
                if 'order_id' not in zone_data.columns:
                    zone_data['order_id'] = zone_data.index.astype(str)
                total_m3 = zone_data['m3'].sum()
                avg_distance = zone_data['Distance'].mean()
                zone_cost_before = zone_data['Cost_Before_Pooling'].sum()

                print(f"  Month {month} | Concrete {concrete_type} | Zone {zone_num} (~{avg_distance:.1f} km): Orders={len(zone_data)}, Total_m3={total_m3:.2f}")

                # Solve for this zone
                if use_ilp and PULP_AVAILABLE:
                    try:
                        assignments, zone_cost_after = solve_zone_ilp(zone_data, TRUCK_SIZES, time_limit=ilp_time_limit)
                    except Exception as e:
                        print("  ILP solver failed for zone (falling back to greedy). Exception:", e)
                        assignments = greedy_pack(zone_data)
                        zone_cost_after = sum(tr['total'] * TRUCK_SIZES[tr['size']]['price_per_m3'] for tr in assignments)
                else:
                    assignments = greedy_pack(zone_data)
                    zone_cost_after = sum(tr['total'] * TRUCK_SIZES[tr['size']]['price_per_m3'] for tr in assignments)

                # record each truck route
                n_assign = len(assignments) if assignments else 0
                if n_assign == 0:
                    print("   WARNING: No trucks created for this zone (empty assignments).")
                zone_cost_after_accum = 0.0
                for truck_num, truck in enumerate(assignments, 1):
                    truck_size = truck['size']
                    truck_load = truck['total']
                    cost_after = truck_load * TRUCK_SIZES[truck_size]['price_per_m3']
                    zone_cost_after_accum += cost_after
                    route_counter += 1
                    truck_routes.append({
                        'Month': month,
                        'Route': route_counter,
                        'Concrete_Type': concrete_type,
                        'Distance_Zone': f"{int(round(avg_distance))}km",
                        'Truck_Size': truck_size,
                        'Orders_Count': len(truck['load']),
                        'Total_m3': truck_load,
                        'Utilization': (truck_load / TRUCK_SIZES[truck_size]['capacity']) * 100,
                        'Price_per_m3': TRUCK_SIZES[truck_size]['price_per_m3'],
                        'Cost_Before': zone_cost_before * (truck_load/total_m3) if total_m3>0 else 0,  # proportional split
                        'Cost_After': cost_after,
                        'Route_Savings': (zone_cost_before * (truck_load/total_m3) - cost_after) if total_m3>0 else 0
                    })
                    print(f"    Truck (size {truck_size} m3): load {truck_load:.2f} m3, orders {len(truck['load'])}, cost €{cost_after:.2f} (util {truck_routes[-1]['Utilization']:.1f}%)")

                zone_savings = zone_cost_before - zone_cost_after_accum
                zone_savings_pct = (zone_savings / zone_cost_before * 100) if zone_cost_before>0 else 0
                print(f"    Zone total - Before: €{zone_cost_before:.2f} | After: €{zone_cost_after_accum:.2f} | Savings: €{zone_savings:.2f} ({zone_savings_pct:.1f}%)")

    # Summaries
    routes_df = pd.DataFrame(truck_routes)
    if routes_df.empty:
        print("\nNo routes produced. Exiting.")
        return

    print("\n=== Fleet Optimization Summary (sample) ===")
    print(routes_df[['Month','Route','Concrete_Type','Distance_Zone','Truck_Size','Orders_Count','Total_m3','Price_per_m3','Utilization','Cost_After','Route_Savings']].to_string(index=False))

    total_cost_before = df['Cost_Before_Pooling'].sum()
    total_cost_after = routes_df['Cost_After'].sum()
    total_savings = total_cost_before - total_cost_after
    total_savings_pct = (total_savings / total_cost_before * 100) if total_cost_before>0 else 0

    print("\n=== Total Optimization Results ===")
    print(f"Total Cost (Before Optimization): €{total_cost_before:.2f}")
    print(f"Total Cost (After Optimization):  €{total_cost_after:.2f}")
    print(f"Total Savings:                    €{total_savings:.2f}")
    print(f"Savings Percentage:               {total_savings_pct:.1f}%")
    print(f"Total Routes Required: {len(routes_df)}")
    print(f"Average Truck Utilization: {routes_df['Utilization'].mean():.1f}%")

    # Truck breakdown
    truck_breakdown = routes_df.groupby('Truck_Size').agg({'Route':'count','Total_m3':'sum','Cost_After':'sum','Route_Savings':'sum'}).reset_index()
    truck_breakdown.columns = ['Truck_Size','Count','Total_m3','Total_Cost','Total_Savings']
    print("\n--- Truck Fleet Breakdown ---")
    print(truck_breakdown.to_string(index=False))

    # Monthly breakdown
    monthly_breakdown = routes_df.groupby('Month').agg({'Route':'count','Total_m3':'sum','Cost_After':'sum','Route_Savings':'sum'}).reset_index()
    monthly_breakdown.columns = ['Month','Routes','Total_m3','Total_Cost_After','Total_Savings']
    # Add before costs
    monthly_breakdown = monthly_breakdown.merge(before_summary[['Month','Total_Cost_Before']], on='Month', how='left')
    monthly_breakdown['Savings_Percent'] = ((monthly_breakdown['Total_Cost_Before'] - monthly_breakdown['Total_Cost_After']) / monthly_breakdown['Total_Cost_Before'] * 100).round(1)
    print("\n--- Monthly Optimization Breakdown ---")
    print(monthly_breakdown[['Month','Routes','Total_Cost_Before','Total_Cost_After','Total_Savings','Savings_Percent']].to_string(index=False))

    # Optionally save results
    routes_df.to_excel("optimized_truck_routes.xlsx", index=False)
    print("\nDetailed routes saved to 'optimized_truck_routes.xlsx' in current directory.")

if __name__ == "__main__":
    print("PuLP available:", PULP_AVAILABLE)
    # Use ILP if available; fallback to greedy if not.
    main(use_ilp=True, ilp_time_limit=60)
```
